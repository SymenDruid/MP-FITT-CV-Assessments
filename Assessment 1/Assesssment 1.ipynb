#### Question 1

Write the code for Local Binary Pattern in python (without using its library function) based on the following steps:

1. Take the input image and convert it into a grayscale image.
2. Use a sliding window technique to traverse the input image.
3. For each pixel location as centre, check its nearest 8 neighbourhood i.e. 3*3 matrix as shown below.
4. Threshold the neighbourhood values based on the value at the centre pixel and generate the 8 bit binary.
5. Convert 8 bit binary into decimal for every pixel as centre

#### Solution

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def pixel_decimal_value(img, center, x, y):

    value = 0
    threshold = img[center[0], center[1]]

    # Top-left
    if img[x-1, y-1] >= threshold:
        value += 1
    # Top
    if img[x-1, y] >= threshold:
        value += 2
    # Top-right
    if img[x-1, y+1] >= threshold:
        value += 4
    # Right
    if img[x, y+1] >= threshold:
        value += 8
    # Bottom-right
    if img[x+1, y+1] >= threshold:
        value += 16
    # Bottom
    if img[x+1, y] >= threshold:
        value += 32
    # Bottom-left
    if img[x+1, y-1] >= threshold:
        value += 64
    # Left
    if img[x, y-1] >= threshold:
        value += 128

    return value

def local_binary_pattern(img):
    height, width = img.shape
    lbp_img = np.zeros((height-2, width-2), dtype=np.uint8) #This will not include the border pixels

    for i in range(1, height-1):
        for j in range(1, width-1):
            lbp_img[i-1, j-1] = pixel_decimal_value(img, (i, j), i, j)

    return lbp_img
```

#### Question 2

Generate the output image using this code for input images given below



#### Solution (a)

```python
# Converting Image into Greyscale
image_path = "image_a.jpg"
input_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply LBP
lbp_result = local_binary_pattern(input_image)

# Display the result using matplotlib
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(input_image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(lbp_result, cmap='gray')
plt.title('LBP Result')
```
